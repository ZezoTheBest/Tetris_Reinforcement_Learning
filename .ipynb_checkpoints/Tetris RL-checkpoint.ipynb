```python
def create_tetris_block(width,height,game_x, game_y):    
    color = np.random.choice(np.array(['cyan','yellow','purple','green','blue','red','orange'])) # color
    #The shape of the block depends on the color
    
    possible = np.arange(game_x, game_x + 11 * width, width)
    x_pos = np.random.choice(possible[2:-5])
    if color == 'yellow': #O block
        rect1 = pygame.Rect(x_pos,game_y,width,height)
        rect2 = pygame.Rect(x_pos + width,game_y,width,height)
        rect3 = pygame.Rect(x_pos, game_y + height, width, height)
        rect4 = pygame.Rect(x_pos + width, game_y+ height, width, height)
    
    elif color == 'cyan': # I block
        rect1 = pygame.Rect(x_pos,game_y,width,height)
        rect2 = pygame.Rect(x_pos + width,game_y,width,height)
        rect3 = pygame.Rect(x_pos + 2 * width, game_y,width, height)
        rect4 = pygame.Rect(x_pos +  3 * width, game_y, width, height)
        
    elif color == 'purple': # T block
        rect1 = pygame.Rect(x_pos,game_y,width,height)
        rect2 = pygame.Rect(x_pos + width,game_y,width,height)
        rect3 = pygame.Rect(x_pos + width, game_y + height, width, height)
        rect4 = pygame.Rect(x_pos +  2 * width, game_y, width, height)
    
    elif color == 'green': # S block
        rect1 = pygame.Rect(x_pos,game_y,width,height)
        rect2 = pygame.Rect(x_pos,game_y + height,width,height)
        rect3 = pygame.Rect(x_pos + width, game_y + height, width, height)
        rect4 = pygame.Rect(x_pos + width, game_y +  2 * height, width, height)
    
    elif color == 'orange' : # L block 
        rect1 = pygame.Rect(x_pos,game_y,width,height)
        rect2 = pygame.Rect(x_pos,game_y + height,width,height)
        rect3 = pygame.Rect(x_pos, game_y +  2 * height, width, height)
        rect4 = pygame.Rect(x_pos + width, game_y +  2 * height, width, height)
    
    elif color == 'red': # J block
        rect1 = pygame.Rect(x_pos,game_y,width,height)
        rect2 = pygame.Rect(x_pos ,game_y + height,width,height)
        rect3 = pygame.Rect(x_pos, game_y + 2 * height, width, height)
        rect4 = pygame.Rect(x_pos - width, game_y +  2 * height, width, height)
    
    elif color == 'blue' : # Z block
        rect1 = pygame.Rect(x_pos,game_y,width,height)
        rect2 = pygame.Rect(x_pos + width,game_y,width,height)
        rect3 = pygame.Rect(x_pos + width, game_y + height, width, height)
        rect4 = pygame.Rect(x_pos +  2 * width, game_y +  height, width, height)
        
    block_rectangles = [rect1,rect2,rect3,rect4]

    return [color, block_rectangles]
    
```

```python
def block_filter(fallen_list):
    for i,block_and_content in enumerate(fallen_list):
        filter_list = [not row in all_ones for row in list(block_and_content[3])]
        block_and_content[1] = [v for (v,b) in zip(block_and_content[1],filter_list) if b]
        block_and_content[3] = [v for (v,b) in zip(block_and_content[3],filter_list) if b]
    
    fallen_list = list(filter(lambda x : len(list(x[3])) > 0, fallen_list))
    return fallen_list

        
```

```python
def remove_row(full_row_numbers,fallen_list, rect_list):
    tetris_block_matrix[full_row_numbers,:] = 0
    
    fallen_list = block_filter(fallen_list)
    rect_list = []
    fallen_list.sort(reverse = True , key = lambda x : np.max(x[3]))    
    for row in full_row_numbers:
        for index,(color,block,counter,row_numbers) in enumerate(fallen_list): 
            for i in range(len(block)):
                if row_numbers[i] != row:
                    rect_list.append((color,block[i],counter))

    return fallen_list,rect_list
                
        
```

```python
def check_full_rows():
    global tetris_block_matrix, score, rect_list

    full_rows = []
    for i in range(tetris_block_matrix.shape[0]):
        if np.all(tetris_block_matrix[i, :] != 0):
            full_rows.append(i)

    if full_rows:
        # Update tetris_block_matrix
        tetris_block_matrix = np.delete(tetris_block_matrix, full_rows, axis=0)
        new_rows = np.zeros((len(full_rows), tetris_block_matrix.shape[1]))
        tetris_block_matrix = np.vstack((new_rows, tetris_block_matrix))

        # Find the highest cleared row
        highest_cleared_row = min(full_rows)

        # Update rect_list
        new_rect_list = []
        for color, rect, counter in rect_list:
            # Check if the block is above the cleared rows
            if rect.y < screen_height * 0.5 - BLOCK_SIZE * 16 + highest_cleared_row * BLOCK_SIZE:
                # Calculate the number of falling rows dynamically
                falling_rows = 1

                # Shift down blocks using a while loop until a collision is detected or reaching the bottom
                while (rect.y + falling_rows * BLOCK_SIZE) + rect.height <= screen_height * 0.5 and not any(
                        (rect.y + falling_rows * BLOCK_SIZE) + rect.height > rect2.y and rect.x == rect2.x
                        for color2, rect2, _ in rect_list if rect2 != rect):
                    falling_rows += 1

                # Apply the final falling_rows adjustment
                rect.y += falling_rows * BLOCK_SIZE
                new_rect_list.append((color, rect, counter))

        # Update rect_list with the modified list
        rect_list = new_rect_list

        score += len(full_rows) * 100  # Increase score for each cleared row
```
