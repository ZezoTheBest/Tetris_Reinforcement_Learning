```python
def create_tetris_block(width,height,game_x, game_y):    
    color = np.random.choice(np.array(['cyan','yellow','purple','green','blue','red','orange'])) # color
    #The shape of the block depends on the color
    
    possible = np.arange(game_x, game_x + 11 * width, width)
    x_pos = np.random.choice(possible[2:-5])
    if color == 'yellow': #O block
        rect1 = pygame.Rect(x_pos,game_y,width,height)
        rect2 = pygame.Rect(x_pos + width,game_y,width,height)
        rect3 = pygame.Rect(x_pos, game_y + height, width, height)
        rect4 = pygame.Rect(x_pos + width, game_y+ height, width, height)
    
    elif color == 'cyan': # I block
        rect1 = pygame.Rect(x_pos,game_y,width,height)
        rect2 = pygame.Rect(x_pos + width,game_y,width,height)
        rect3 = pygame.Rect(x_pos + 2 * width, game_y,width, height)
        rect4 = pygame.Rect(x_pos +  3 * width, game_y, width, height)
        
    elif color == 'purple': # T block
        rect1 = pygame.Rect(x_pos,game_y,width,height)
        rect2 = pygame.Rect(x_pos + width,game_y,width,height)
        rect3 = pygame.Rect(x_pos + width, game_y + height, width, height)
        rect4 = pygame.Rect(x_pos +  2 * width, game_y, width, height)
    
    elif color == 'green': # S block
        rect1 = pygame.Rect(x_pos,game_y,width,height)
        rect2 = pygame.Rect(x_pos,game_y + height,width,height)
        rect3 = pygame.Rect(x_pos + width, game_y + height, width, height)
        rect4 = pygame.Rect(x_pos + width, game_y +  2 * height, width, height)
    
    elif color == 'orange' : # L block 
        rect1 = pygame.Rect(x_pos,game_y,width,height)
        rect2 = pygame.Rect(x_pos,game_y + height,width,height)
        rect3 = pygame.Rect(x_pos, game_y +  2 * height, width, height)
        rect4 = pygame.Rect(x_pos + width, game_y +  2 * height, width, height)
    
    elif color == 'red': # J block
        rect1 = pygame.Rect(x_pos,game_y,width,height)
        rect2 = pygame.Rect(x_pos ,game_y + height,width,height)
        rect3 = pygame.Rect(x_pos, game_y + 2 * height, width, height)
        rect4 = pygame.Rect(x_pos - width, game_y +  2 * height, width, height)
    
    elif color == 'blue' : # Z block
        rect1 = pygame.Rect(x_pos,game_y,width,height)
        rect2 = pygame.Rect(x_pos + width,game_y,width,height)
        rect3 = pygame.Rect(x_pos + width, game_y + height, width, height)
        rect4 = pygame.Rect(x_pos +  2 * width, game_y +  height, width, height)
        
    block_rectangles = rect1,rect2,rect3,rect4  

    return color, block_rectangles 
    
```

```python

tetris_block_matrix = np.zeros((30,10))
pygame.init() # Initialize game
screen = pygame.display.set_mode((0,0), pygame.FULLSCREEN) # Set to full screen

#==========================================================================
pygame.display.set_caption('Tetris') # Set title

#Set icon
icon = pygame.image.load('Tetris.png')
pygame.display.set_icon(icon) 

clock = pygame.time.Clock()

#Screen width and height
screen_width = pygame.display.Info().current_w
screen_height = pygame.display.Info().current_h


#========================================================================
#Put the background
main_bg = pygame.image.load('tetris_bg.jpg').convert()
main_bg = pygame.transform.rotate(main_bg,90) # rotate pic by 90 degrees
scale_factor_height = screen_height / main_bg.get_rect().height #fit pic height
scale_factor_width = screen_width / main_bg.get_rect().width #fit pic width

main_bg = pygame.transform.scale(main_bg,
                                 ((scale_factor_width * main_bg.get_rect().width),(scale_factor_height * main_bg.get_rect().height)) )

#========================================================================
#Scoreboard
score = 0
score_screen = pygame.surface.Surface((200,300))
#Set score font
score_font = pygame.font.Font('tetris_font.ttf', 25)
#Set score text
score_word = score_font.render('SCORE',False,'White')
score_value = score_font.render(str(score),False,'White')

#====================================================================
#Tetris Game Screen

'''The game is 40 blocks high and 10 blocks wide'''
BLOCK_SIZE = 22
tetris_game = pygame.surface.Surface((BLOCK_SIZE * 10, BLOCK_SIZE * 30))

#===============================================================================
counter = 1 # Number of blocks up until now
collision_detected = False # Detects collision with another block
color, tetris_block = create_tetris_block(
            BLOCK_SIZE , BLOCK_SIZE, screen_width *0.5 - BLOCK_SIZE * 5, screen_height * 0.5 - BLOCK_SIZE * 16) # create a tetris block

rect_list = [] # Saves all rectangles to keep them displayed


move_left = False # Left arrow detector

move_right = False # right arrow detector

rotate_clockwise = False # rotate clockwise detector

rotation_orientation = 0

while True: # While loop for game
    for event in pygame.event.get():
        if event.type == pygame.QUIT: # If we exit, close the game and exit the while loop
            pygame.quit()
            exit()
        elif event.type == pygame.KEYDOWN: # If we press escape(Esc), we exit the game
            if event.key == pygame.K_ESCAPE:
                pygame.quit()
                exit()
            elif event.key == pygame.K_LEFT:
                move_left = True
            elif event.key == pygame.K_RIGHT:
                move_right = True
            elif event.key == pygame.K_RETURN or event.key == pygame.MOUSEBUTTONDOWN :
                rotate_clockwise = True
            
        else:
            move_left = False
            move_right = False
            rotate_clockwise = False
            
    #================================
    screen.blit(main_bg,(0,0)) # add main background
    
    score_screen.blit(score_word,(40,20)) # add score text and score to score surface
    score_screen.blit(score_value,(90,100))
    screen.blit(score_screen, (screen_width * 0.85,screen_height * 0.5 - 150)) # add score surface to main surface
    score_value = score_font.render(str(score),False,'White')

    
    #Draw vertical lines
    for x in range(0, tetris_game.get_rect().width + BLOCK_SIZE, BLOCK_SIZE):
        pygame.draw.line(tetris_game, 'White', (x, 0), (x, tetris_game.get_rect().height))

    # Draw horizontal lines
    for y in range(0, tetris_game.get_rect().height + BLOCK_SIZE, BLOCK_SIZE):
        pygame.draw.line(tetris_game, 'White', (0, y), (tetris_game.get_rect().width, y))
    screen.blit(tetris_game,(screen_width *0.5 - BLOCK_SIZE * 5,screen_height * 0.5 - BLOCK_SIZE * 16)) # Tetris game
    
    if collision_detected:
        counter += 1
        rotation_orientation = 0
        color, tetris_block = create_tetris_block(
            BLOCK_SIZE , BLOCK_SIZE, screen_width *0.5 - BLOCK_SIZE * 5, screen_height * 0.5 - BLOCK_SIZE * 16) # create a tetris block


    for rect in tetris_block: # Add the rectangles to a list to remember it 
        rect_list.append((color,rect,counter,int((rect.y -screen_height * 0.5 - BLOCK_SIZE * 16) // rect.height)))
    
        #draw tetris block
    for color,rect,counter,row in rect_list:
        pygame.draw.rect(screen,color,(rect.x,rect.y,rect.width,rect.height))
    

    #row number of the tetris blocks 
    collision_detected, rotate_success = move_block(tetris_block,color,rotation_orientation,move_left,move_right,rotate_clockwise,
               screen_width *0.5 - BLOCK_SIZE * 5, screen_height * 0.5 - BLOCK_SIZE * 16,counter)
    if rotate_success:
        rotation_orientation = (rotation_orientation + 1) % 4
    
    all_ones = [row_number for row_number in range(tetris_block_matrix.shape[0]) if np.all(tetris_block_matrix[row_number,:] != 0)]
    if collision_detected and len(all_ones) != 0:
        rect_list = remove_row(all_ones,rect_list)
        score += 100


    pygame.display.update() # update
    clock.tick(5) # Run at 60 FPS
    
    
```
